```python
dummy=pd.get_dummies(df[['employment','education','married_status','x','life','color','gender','credit_score','place','amount']])
```

```python
df1=pd.concat([df,dummy],axis=1)
```

```python
dfe=df1.drop(['employment','education','married_status','x','life','color','gender','credit_score','place','amount'],axis=1)
```

```python
dfe
```

```python
dfe['y'].value_counts()
```

```python
dfe['z'].value_counts()
```

```python
dfe.drop(['y','z'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(dfe)
x=scaler.transform(dfe)
x
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=2,init='k-means++',random_state=42)
y_model=model.fit_predict(x)
y_model
```

```python
dfe['cluster']=y_model
```
